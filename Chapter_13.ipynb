```python
import numpy as np 
import matplotlib.pyplot as plt
import urllib, pandas 
import scipy.stats as st
url  = "http://www.stat.cmu.edu/~larry/all-of-statistics/=data/carmileage.dat"

file = urllib.request.urlopen(url)


n = 82 
lines = []
line_count = 0 
for line in file:
    line_count += 1
    if (line_count >= 30):  
        full_line = line.decode("utf-8").split() 
        lines.append(full_line)

X = []
for line in lines: 
    X_line = [float(num) for num in line[-5:]]
    X.append(np.asmatrix(X_line))
    
X = np.concatenate(X, axis=0)


Y = X[:,[2]]
X = X[:,[0,1,3,4]]

# ==================== PART (A) ====================
# determine beta vector
XTX_inv = np.linalg.inv(np.matmul(X.transpose(),X))
beta = np.matmul(np.matmul(XTX_inv, X.transpose()),Y)

# estimate variance 
k = len(beta)
eps = np.matmul(X, beta)-Y
sig2 = 1/(n-k)*np.sum(np.multiply(eps, eps))

# estimate variance for vector of parameter
var = sig2 * XTX_inv

#Display result 
names = ['VOL','HP','SP','WT']

id2wald = {}

print("Part (a) - Multiple linear regression model analysis\n")
print(f'{"Name":>6} | {"Parameter":>10} | {"std err.":>10} | {"t-value":>10} | {"p-value":>10}')
print(f'{"---------------------------------------------------------------"}')
for i in range(0,k):
    name = names[i]
    param = float(beta[i]) 
    std_err = np.sqrt(float(var[[i],[i]]))
    t_val = (param - 0)/std_err
    p_val = 2*st.norm.cdf(-abs(t_val))
    print(f'{name:>6} | {param:10.4f} | {std_err:10.6f} | {t_val:10.6f} | {p_val:10.5f}')
    # build map from cov. name to Wald stats for part (C) 
    id2wald[abs(t_val)] = i
    

# ==================== PART (B) ====================
print('\n\nPart (b)')
print('   (i) Model selection using forward stepwise regression and AIC')
cov_remaining = [0,1,2,3]  # set of covariates that are kept
AIC_max = float("-inf")
S = set([])
while True: 
    AIC = []
    for i in cov_remaining: 
        
        # build model using covariates in set S + covariate i  
        S.add(i)
        X_i = X[:,list(S)]
        beta_i = np.matmul(np.linalg.inv(np.matmul(X_i.transpose(),X_i))*X_i.transpose(), Y) 
        sigma2_hat = 1/(n-len(S)) * np.sum(np.power((np.matmul(X_i, beta_i) - Y),2))

        # compute AIC if covariate `i` is included  
        log_likelihood = 0
        for j in range(n): 
            f = st.norm.pdf(x = Y[j], loc = X_i[j]* beta_i, scale = sigma2_hat)
            log_likelihood += np.log(f)
        
        AIC.append(log_likelihood - len(S))  
        S.discard(i)
    

    print("AIC table: ", end="")
    for k in range(len(cov_remaining)): 
        print(f'{float(AIC[k]):5.2f} ({names[cov_remaining[k]]}), ', end = "")

    if (max(AIC) > AIC_max or len(cov_remaining) == 0): 
        AIC_max = max(AIC)
        i_max = AIC.index(max(AIC))
        print(f'-> Keeping {names[cov_remaining[i_max]]} increases AIC to {float(max(AIC)):.2f}')
        id = cov_remaining[i_max]
        cov_remaining.remove(id)
        S.add(id)
    else: 
        print("-> AIC cannot be increased. ")
        break 

    
    
    
    
print('\n   (ii) Model selection using backward stepwise regression and AIC')
cov_kept = [0,1,2,3]  # set of covariates that are kept
AIC_max = float("-inf")
S = set(cov_kept) # set of cov removed 
while True: 
    AIC = []
    for i in cov_kept: 
        # build model using covariates in set S + covariate i  
        S.remove(i)
        X_i = X[:,list(S)]
        beta_i = np.matmul(np.linalg.inv(np.matmul(X_i.transpose(),X_i))*X_i.transpose(), Y) 
        sigma2_hat = 1/(n-len(S)) * np.sum(np.power((np.matmul(X_i, beta_i) - Y),2))

        # compute AIC if covariate `i` is included  
        log_likelihood = 0
        for j in range(n): 
            f = st.norm.pdf(x = Y[j], loc = X_i[j]* beta_i, scale = sigma2_hat)
            log_likelihood += np.log(f)
        AIC.append(log_likelihood - len(S))  
        S.add(i)
    

    print("AIC table: ", end="")
    for k in range(len(cov_kept)): 
        print(f'{float(AIC[k]):5.2f} ({names[cov_kept[k]]}), ', end = "")

    if (max(AIC) > AIC_max or len(cov_kept) == 0): 
        AIC_max = max(AIC)
        i_max = AIC.index(max(AIC))
        print(f'-> Removing {names[cov_kept[i_max]]} increases AIC to {float(max(AIC)):.2f}')
        id = cov_kept[i_max]
        cov_kept.remove(id)
        S = set(cov_kept)
    else: 
        print("-> AIC cannot be increased. ")
        break 

        
# ==================== PART (C) ====================
print('\n\nPart (c) - Model selection using Zheng-Loh\n')


# ids = column ids in decresing order of wald statistic. 
# e.g. ids[0] = column position with the largest wald statistic 
ids = [] 
for i in sorted(id2wald.items())[::-1]:
    ids.append(i[1])

for i in range(len(ordered_ids)): 
    X_i = X[:,ids[0:(i+1)]]
    beta_i = np.matmul(np.linalg.inv(np.matmul(X_i.transpose(),X_i))*X_i.transpose(), Y) 
    RSS = np.sum(np.power((np.matmul(X_i, beta_i) - Y),2)) 
    sigma2_hat = 1/(n-len(S)) * RSS
    print(f'RSS(j) + j*sig2*log(n) =  {RSS + (i+1)*sigma2_hat*np.log(n):.3f}, if {{', end="")
    for p in ids[0:(i+1)]: 
        print(f'{names[p]} ', end="")
    print(f'}} is chosen. (i.e. j = {i+1})')

# ==================== PART (D) ====================
print('\n\nPart (d) - All regressions\n')

# generate all 2^4 - 1 possible sets of covariates 
sets = []
for i in range(1,16): 
    bin_arr_char = (list(bin(i))[2:])[::-1]    
    subset_ids = []
    for i in range(len(bin_arr_char)): 
        if (bin_arr_char[i] == '1'): 
            subset_ids.append(i)
    sets.append(subset_ids)
    

BIC = []
C_p = []
subset_names = []
for subset in sets: 
    S = len(subset)
    X_i = X[:,subset]
    beta_i = np.matmul(np.linalg.inv(np.matmul(X_i.transpose(),X_i))*X_i.transpose(), Y)
    RSS = np.sum(np.power((np.matmul(X_i, beta_i) - Y),2)) 
    sigma2_hat = 1/(n-S) * RSS

    # compute BIC when S = subset  
    log_likelihood = 0
    for j in range(n): 
        f = st.norm.pdf(x = Y[j], loc = X_i[j]* beta_i, scale = sigma2_hat)
        log_likelihood += np.log(f)
    BIC.append(float(log_likelihood - 1/2*S*np.log(n)))
    
    # compute Mallow's C_p statistic 
    C_p.append(RSS + 2*S*sigma2_hat) 
    
    # generate set name for display 
    subset_name = "" 
    for i in subset: 
        subset_name += (names[i] + ", ")
    subset_names.append(subset_name[:-2])

    
# display results 

print(f'{"covariates":>15} | {"BIC":>10} | {"C_p":>10}')
print(f'{"---------------------------------------------------"}')
for i in range(0,len(BIC)):
    print(f'{subset_names[i]:>15} | {BIC[i]:10.1f} | {C_p[i]:10.1f}')


# generate visualization 

dic = {}
for i in range(len(BIC)): 
    dic[C_p[i]] = BIC[i]

C_p_sorted =[]
BIC_sorted = sorted(dic)
for bic in BIC_sorted: 
    C_p_sorted.append(dic[bic])


fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(BIC_sorted, 'g.')
ax2.plot(C_p_sorted, 'b.')
ax1.set_ylabel('BIC', color='g')
ax2.set_ylabel('C_p', color='b')
plt.title('Relationship between BIC and C_p (sorted w.r.t BIC)')
plt.show()    

```
