### Introductory terms: 
<b> Sample space </b> = set all of all the possible outcome in a random experiement. <br> 
<b> Sample outcome, Elements, Realization </b> = Single elements in the sample space.<br> 
<b> Event </b> = set within the sample sample <br>

### Axioms of Probability: 
A function can be considered to be a probability function if: <br>
$P(A) \ge 0$ for every A<br>
$P(\Omega) = 1$ <br>
If A and B are disjoint, then $P(A\cup B) = P(A) + P(B)$ <br> 

With these 3 axioms, multiple lemmas can be generated . 

### Independance of Events: 
Two events $A$ and $B$ are considered to be independant IFF $P(AB) = P(A)P(B)$ <br>
If two events are disjoint (i.e. cannot occur simultaneously), then they cannot be independant. <br>


### Conditional Probability: 
If $P(B) > 0$, then the conditional probability of $A$ given $B$ is: <br>
$P(A\mid B) = \frac{P(AB)}{P(B)}$


### Law of total probability 
The probability of an event $B$ can expressed as the sum of the conditional probabilty of $B$ over partitions $A_{i}$ where $\Omega = \sum_{i=1}^{k}{A_{i}}$. This gives us $P(B) = \sum_{i=1}^{k}{P(B\mid A_{i})P(A_{i})}$

```python

import random 
import matplotlib.pyplot as plt
import numpy as np 



def trial(n, p):
    a = 0
    for i in range(0,n):
        if (random.random() < p):
            a +=1
    return a 

trial_size = 1000

p = 0.3
n_values = [10, 100, 1000]

for n in n_values: 
    N = np.arange(0,trial_size,1)
    X_mean = np.empty([trial_size,1])
    sum = 0 
    for i in range(0,trial_size):
        sum += trial(n, p) 
        X_mean[i][0] = sum / (i+1)

    plt.figure(1)
    plt.plot(N,X_mean)
    plt.title("Binomial distribution sampling simulation for p = " + str(p) + ", n = " + str(n))
    plt.plot([0, trial_size],[n*p, n*p])
    plt.xlabel("Number of trial")
    plt.ylabel("mean X")
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib as mtp
import random


def contains(set,val): 
    for elem in set: 
        if (val == elem):
            return True
    return False

n = 1000 
A = (3,4,6)
B = (1,2,3,4)

PA_sum = 0
PB_sum = 0
PAB_sum = 0 

N = np.arange(0,n,1)
residual = np.empty(N.shape)

for i in range(0,n): 
    X = np.random.randint(1,7)
    if (contains(A,X)): 
        PA_sum += 1
    if (contains(B,X)): 
        PB_sum += 1
    if (contains(B,X) & contains(A,X)): 
        PAB_sum += 1
    residual[i] = PAB_sum / (i+1) - ((PA_sum / (i+1)) * (PB_sum / (i+1)))
        

plt.figure(1)
plt.title("Residual $\mathbb{\hat{P}}(AB) -\mathbb{\hat{P}}(B)\mathbb{\hat{P}}(A) $")
plt.xlabel("Number of trial")
plt.plot(N, residual)
plt.show()

```
