### Definition of Expectation 

Unifying definition of the expected value of a random variable


$$
 \mathbb{E}(x) = \mathbb{E}X = \mu_X = \int_{-\infty}^{\infty} f(x) dF(x) = 
  \begin{cases} 
      \hfill \sum_{\omega}xf(x)   \hfill & \text{for discrete RVs} \\
      \hfill  \int_{\omega} xf(x) dx \hfill & \text{for continuous RVs} \\
  \end{cases}
$$

Properties of expected value: <br>
$\sum_{i}\mathbb{E}(a_iX_i) = \sum_{i}a_i\mathbb{E}(X_i)$ <br>
$\mathbb{E}(\Pi_i(X_i)) = \Pi_i(\mathbb{E}(X_i)) \quad$  Given ($X_is$ are independant!)<br> 


### Lazy Statistician law

given $Y = r(X)$ <br>
$$\mathbb{E}(Y) = \mathbb{E}(r(x)) = \int r(x)dF(x)$$
The law of the Lazy Statistician can be useful when computing the expected value of a transformed random variable because it allows to bypass the process of finding the pdf of the transformed RV. 


### Moment of Random variable 

The $k^{th}$ moment of a random variable is defined as $\mathbb{E}(X^k)$ assuming $\mathbb{E}(|X|^k) < \infty$

If j < k and the $k^{th}$ moment of random variable exists, then its $j^{th}$ moment also exists.  

### Variance Definition 

Variance is defined as follows: 

$$\mathbb{V}(X) = \mathbb{E}((X-\mu_X)^2)$$

properties of variance: <br>
$\mathbb{V}(X) = \mathbb{E}(X^2) - \mu_x^2$ <br>
$\sum_{i}\mathbb{V}(a_iX_i) = \sum_{i}a_i^2\mathbb{V}(X_i) \quad$ Given ($X_is$ are independant!)<br> 
$\mathbb{V}(aX + b) = a^2\mathbb{V}(X)$

### Sampling Random Variable

Given a random experiment sample of size n, where $X_i$s are i.i.d. the <b>sample mean</b> $\bar{X}_n$ and <b>sample variance</b> $S_n$ are defined as follows: <br>

$$\bar{X}_n = \dfrac{1}{n}\sum_i^n X_i \quad\quad S_n =  \dfrac{1}{n-1}\sum_i^n(X_i - \bar{X})^2$$

Properties of sampling random variables: <br>
$\mathbb{E}(\bar{X}_n) = \mathbb{E}(X)$ <br>
$\mathbb{V}(\bar{X}_n) = \dfrac{1}{n}\mathbb{V}(X)$ <br> 
$\mathbb{E}(S_n^2) = \mathbb{V}(X)$

### Covariance and Correlation Definition 

The <b>covariance</b> $Cov(X,Y)$ and <b>correlation</b> $\rho(X,Y)$ of two random variables X and Y are defined as follow: 

$$Cov(X,Y) = \mathbb{E}((X-\mu_X)(Y-\mu_Y)) \quad\quad \rho(X,Y) = \dfrac{Cov(X,Y)}{\sigma_x\sigma_y}$$

propertie of covariance and correlation: <br> 
$Cov(X,Y) = \mathbb{E}(XY) - \mathbb{V}(X)\mathbb{V}(Y)$ <br>
$ -1 \le \rho(X,Y) \le 1$ <br>
$X$ and $Y$ are independant $\implies \rho(X,Y) = 0$ <br> 
$Y = aX + b \implies \rho(X,Y) = sign(a) $


Other properties of variance: <br>
$\mathbb{V}(X + Y) = \mathbb{V}(X) + \mathbb{V}(Y) + 2Cov(X,Y)$, which can be generalized to :<br> 
$\mathbb{V}(a_i\sum_i X_i) = \sum_ia_i^2\mathbb{V}( X_i)  + 2\sum_{i<j}a_ia_jCov(X_i, Y_i)$ <br>   

### Covariance for Vectors of Random Variables


Given a vector of RVs of size nx1 $X = [X_1, X_2, ... X_n]^T$, we have: <br> 

The mean of the RV vector defined as: <br> 
$\mathbb{E}(X) = [\mathbb{X_1}, \mathbb{X_2}...,  \mathbb{X_n}]^T$

The <b>Variance-Covariance matrix</b> $\sum$ of the vector defined as : 

$
\sum = 
\begin{bmatrix}
    \mathbb{V}(X_1) & Cov(X_1,X_2) & Cov(X_1, X_3) & \dots  & Cov(X_1, X_n) \\
    Cov(X_2,X_1) & \mathbb{V}(X_2) & Cov(X_2, X_3) & \dots  & Cov(X_2, X_n) \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    Cov(X_n,X_1) & Cov(X_n,X_2) & Cov(X_n, X_3) & \dots  & \mathbb{V}(X_n)
\end{bmatrix}
$

```python
import numpy as np
import matplotlib.pyplot as plt

N = 10000
X_normal_sum = 0
X_cauchy_sum = 0

X_n_normal = np.empty([N,1])
X_n_cauchy = np.empty([N,1])
n = np.empty(X_n_normal.shape)

for i in range(0,N):
    n[i] = i+1
    X_normal_sum += np.random.normal()
    X_n_normal[i] = X_normal_sum / (i+1)
    X_cauchy_sum += np.random.standard_cauchy()
    X_n_cauchy[i] = X_cauchy_sum / (i+1)

plt.figure(1)
plt.plot(n,X_n_normal)
plt.plot(n,X_n_cauchy)
plt.title(r"Variation of $\bar{X}_n$ as trial number increases for Normal and Cauchy distributions")
plt.legend([r"$\bar{X}_n$ Normal dist.",r"$\bar{X}_n$ Cauchy dist."])
plt.xlabel("trial number")
plt.ylabel(r"$\bar{X_n}$")
plt.show()
```

<b>Gamma Distribution</b> <br>
$
\mathbb{E}(X)

<b>Beta Distribution</b> <br>

$
\mathbb{E}(X)

### Exercise 3.21

Prior to computing $cov(X,Y)$, two results can be derived: <br>  
(1) Computing the expectation on both sides of $\mathbb{E}(Y\mid X) = X$ and applying the law of iterated expectation results in: <br>
$\mathbb{E}(Y\mid X) = X \implies \mathbb{E}(\mathbb{E}(Y\mid X)) = \mathbb{E}(X) \implies \mathbb{E}(Y) = \mathbb{E}(X)$ <br> 
(2) Applying the law of iterated expected on $\mathbb{E}(XY)$ results in:<br> 
$\mathbb{E}(YX) = \mathbb{E}(\mathbb{E}(YX\mid X)) = \mathbb{E}(X\mathbb{E}(Y\mid X)) =  \mathbb{E}(X^2)$ 

Using both result into the definition of covariance leads to: <br> 
$cov(X,Y) = \mathbb{E}(YX) - \mathbb{E}(X)\mathbb{E}(Y) =  \mathbb{E}(X^2) - \mathbb{E}(X)^2 = \mathbb{V}(X)$ <br>

### Exercise 3.24

The moment generating function of exponential functions is: <br>
$\mathbb{E}(Xt) = \int\limits_0^{\infty}\dfrac{e^{xt}}{\beta}e^{-x/\beta} 
= \int\limits_0^{\infty}\dfrac{1}{\beta}e^{-x/\beta + tx}
= \int\limits_0^{\infty}\dfrac{(1/\beta - t)}{\beta(1/\beta - t)}e^{-x(1/\beta - t)}
= \dfrac{1}{\beta(1/\beta - t)}
= \dfrac{1}{1 - t\beta} \text{  for } t < 1/\beta
$ <br>

Let $Y = \sum_{i=1}^n X_i$ <br> 
Since $X_i$s are iid, $Y$'s moment generating function can be obtained as follows: 
$\psi_{Y}(t) = \Pi_{i=1}^n\psi_X(t) =\big(\dfrac{1}{1 - t\beta}\big)^n$ <br> 
which corresponds to the moment generating function of a gamma function $Y$ ~ $Gamma(n,\beta)$

```python
import sympy as sym

# problem 3.13
x = sym.Symbol('x')
y = sym.Symbol('y')
z = sym.Symbol('z')

# 1. Three definite integral need to be computed in order to obtain the CDF of Z

# part one (for 8<t<10)
F3 = sym.integrate(sym.integrate((x+y)/3,(y,(2*x+8-z)/3,2)),(x,(z-8)/2,1)) + sym.integrate(sym.integrate((x+y)/3,(y,0,2)),(x,0,(z-8)/2))
F3 = sym.simplify(F3)
# part two (for 4<t<8)
F2 = sym.integrate(sym.integrate((x+y)/3,(y,(2*x+8-z)/3,2)),(x,0,1))
F2 = sym.simplify(F2)
# part three (for 2<z<4)
F1 = sym.integrate(sym.integrate((x+y)/3,(y,(2*x+8-z)/3,2)),(x,0,(z-2)/2))
F1 = sym.simplify(F1)
print("Computed CDFs of Z")
print(F1)
print(F2)
print(F3)

# 2. Take the derivate of CDFs to obtain PDFs
print("Computed pdf of Z")
f1 = sym.diff(F1,z)
f2 = sym.diff(F2,z)
f3 = sym.diff(F3,z)
print(f1)
print(f2)
print(f3)

# 3. Compute the mean
mean = sym.integrate(f1*z,(z,2,4)) + sym.integrate(f2*z,(z,4,8)) + sym.integrate(f3*z, (z,8,10))
print("mean = " + str(mean))

# 4. Compute the Variance
Variance = sym.integrate(f1*(z-mean)**2,(z,2,4)) + sym.integrate(f2*(z-mean)**2,(z,4,8)) + sym.integrate(f3*(z-mean)**2, (z,8,10))
print("Variance = " + str(Variance))
```
