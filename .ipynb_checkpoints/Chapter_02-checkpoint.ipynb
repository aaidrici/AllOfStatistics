Random variable definition: Mapping between an element of a sample space and a real value

Cumulative distribution function (CDF) is a function $F_x:\mathbb{R} \rightarrow [0-1]$ defined by $F_x(x) = P(X \le x)$


A random variable is discrete if it takes countably many values. We define the <b>probability function</b> or <b>probability mass function (pmf)</b> for X by  $f_X(x)$ as $P(X = x)$   
    
A random variable is continuous if there exists a function $f_X(x)$ such that $\int_{-\infty}^{\infty}f_X(x) = 1$, $f_X(x)\ge0$ for all x, and for every $a\le b$,  $\int_{-\infty}^{\infty}f_X(x)dx = P(a\le x\le b)$. The function is called <b> probability density function (pdf)</b>. We have that $F(X) = \int_{-\infty}^{x}f_X(x)dx$ and $P(a\le x\le b) = \int_{a}^{b}f_X(x)dx$ and also $f_X(x) = F'(x)$


Lemma 1.25: Given the CDF $F$ of a random variable $X$, we have that: 
1. $P(X = x) = F(x) - F(x^{-})$ where $F(x^-) = \lim_{y\to x}F(y)$
2. $P(x < X \le y) = F(y) - F(x)$
3. $P(X > x) = 1 - F(x)$
4. if $X$ is continuous, then $F(b) - F(a) = P(a<X<b) = P(a\le X\le b) = P(a<X\le b) = P(a\le X<b)$

### Transformation of Random Variables 

#### Discrete Case
Let $Y = r(X)$. Given $X$ is discrete, $\mathbb{P}(Y=y) = \mathbb{P}(r(X)=y) = \mathbb{P}(\{x : r(x)=y\})$ <br> 
In other words, for every possible value of Y, find all values of X such that $r(X) = Y$ and attribute the sum of their probability to P(Y = y).

#### Continous Case (approach 1)
Let $Y = r(X)$. Given $X$ is continuous, the following approach can be used to obtain $Y$'s pdf: <br> 
1. Find the set $A_y = \{x : r(X) \leq y\}$
2. Compute $F_y(y) = \mathbb{P}(Y<y) = \mathbb{P}(r(X) < y) = \int_{A_y}f_X(x)dx $
3. $f_y(y) = F_Y'(y)$

#### Continous Case (approach 2)
Alternatively, if the transformation function $y = r(x)$ is monotone increasing or decreasing, then its inverse function $x = s(y)$ exists and $Y$'s pdf can be obtained as follows: <br> 

$$f_Y(y) = f_X(s(y))\mid\dfrac{ds(y)}{dy}\mid$$

From the definition of random variable independance, we know that two RVs $X$ and $Y$ are independant if: <br>
$\mathbb{P}(X\in A, Y\in B) = \mathbb{P}(X\in A)\mathbb{P}(Y\in B) \iff$ <br>
$\sum\limits_{X \in A}\sum\limits_{Y\in B}\mathbb{P}(X = x, Y =y ) = \sum\limits_{x\in A}\mathbb{P}(X = x)\sum\limits_{y \in B}\mathbb{P}(Y =y) = \sum\limits_{x\in A}\sum\limits_{y \in B}\mathbb{P}(X = x)\mathbb{P}(Y =y) \iff $ <br> 
$\mathbb{P}(X = x, Y =y ) = \mathbb{P}(X = x)\mathbb{P}(Y =y) \quad\forall x,y\iff$ <br>
$f_{X,Y}(x,y) = f_{X}(x)f_{Y}(y) \quad\forall x,y$

```python
import numpy as np 
import matplotlib.pyplot as plt 

def f(y): 
    return 1/np.sqrt(2*np.pi)*y**(-np.log(y)/2-1)

# PDF plot 
Y = np.arange(0.001,8,0.01)
X = np.empty(Y.shape)
for i in range(0,len(X)): 
    X[i] = f(Y[i])
plt.figure(1)
plt.plot(Y,X)
plt.xlabel('Y')
plt.ylabel('f(y)')
plt.title(r'Graph 2.13a: $f(y) = (1/\sqrt{2\pi})y^{-ln(y)/2-1}$')
plt.show()

# Histogram plot
N_points = 10000
n_bins = 100
Y2 = np.random.randn(N_points)
for i in range(0,len(Y2)): 
    Y2[i] = np.exp(Y2[i])

# We can set the number of bins with the `bins` kwarg
#axs[0].hist(X2, bins=n_bins)
plt.hist(Y2, bins=n_bins, range=[Y[0],Y[-1]])
plt.title(r"Graph 2.13b: Histogram of $f(y)$")
plt.show()
```


### Exercise 2.19

Given $Y = r(X)$ and $X = s(Y)$ where $(X)$ is: 
1. Strictly monotone increasing: <br> 
$F_Y(y) = \mathbb{P}(Y \leq y) =  \mathbb{P}(r(X) \leq y) = \mathbb{P}(X \leq s(y)) = F_X(s(y))$ <br> 
Taking the derivative of $y$ on both sides and applying the chain rule leads to: <br> 
$f_y(y) = \dfrac{dF_X(s(y))}{dy} = f_X(s(y))\dfrac{ds(y)}{dy} \quad\text{where }\dfrac{ds(y)}{dy} > 0 $

2. Strictly monotone decreasing: <br> 
$F_Y(y) = \mathbb{P}(Y \leq y) =  \mathbb{P}(r(X) \leq y) = \mathbb{P}(X \geq s(y)) = 1-\mathbb{P}(X < s(y)) = 1-F_X(s(y))$ <br> 
Taking the derivative of $y$ on both sides and applying the chain rule leads to: <br> 
$f_y(y) = \dfrac{dF_X(s(y))}{dy} = -f_X(s(y))\dfrac{ds(y)}{dy} \quad\text{where }\dfrac{ds(y)}{dy} < 0 $

The only difference between the two cases is the need to force $ds(y)/dy$ to be positive (by multiplying it to $-1$) when r(X) is monotone strictly decreasing. Putting $ds(y)/dy$ in absolute value allows to use the same equation for both cases.
