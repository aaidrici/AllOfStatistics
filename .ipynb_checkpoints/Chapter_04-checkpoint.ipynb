# Chapter 4 Notes


inequalities in probabilities are useful because they are applicable to all pmf and pdf, provided they meet the requiements associated to the inequality. 

### Markov's inequality 
$\mathbb{P}(X > t) \le \dfrac{\mathbb{E}(X)}{t} \quad \forall t > 0$ provided $X \ge 0$


### Chebychev's inequality 
$\mathbb{P}(\mid X - \mu\mid \ge t) \le \dfrac{\sigma^2}{t^2}$ <br>


### Hoeffding's Inequality
Given $Y_i$'s are independant, $\mathbb{E}(Y_i) = 0$ and $a_i\le Y_i\le b_i$, then for any $t>0$, we have that: 

$\mathbb{P}(\sum_{i=1}^n Y_i \ge \epsilon) \le e^{-t\epsilon}\Pi_{i=1}^ne^{t^2(b_i-a_i)^2/8}$

### Theorem 4.5
Given $X_i$'s ~ Bernoulli($p$), then for any $\epsilon > 0$: <br>
$\mathbb{P}(\mid \bar{X_n} - p \mid > \epsilon) \le 2e^{-2n\epsilon^2}$


### Mill's inequality 
Applicable to normal distribution: $Z$ ~ $N(0,1)$ <br>
$\mathbb{P}=(\mid Z \mid > t) \le \sqrt{\dfrac{2}{\pi}}\dfrac{e^{-t^2/2}}{t}$


### Cauchy-Schwartz inequality
Given $X$ and $Y$ have a finite variance: <br>
$\mathbb{E}\mid YX\mid \le \sqrt{\mathbb{E}(X^2)\mathbb{E}(Y^2)}$


### Jensen's inequality 

$\mathbb{E}(g(x)) \ge g(\mathbb{E}(x))$ provided g(x) is convex. <br>
$g(\mathbb{E}(x)) \ge \mathbb{E}(g(x))$ provided g(x) is concave.

 = \mathbb{P}(\mid X - \mu\mid \ge k\sigma) \le \dfrac{1}{k^2}
$ <br>

$\mathbb{P}(\mid\bar{X_n}-p \mid \ge \epsilon) \le \dfrac{\sigma_{\bar{X}}^2}{\epsilon^2} =\dfrac{\sigma_X^2/n}{\epsilon^2} = \dfrac{p(1-p)}{n\epsilon^2}$


<b> Bounding using Hoeffding's inequality: </b> <br>
Recall Hoeffding's inequality is only applicable for a sum of random variables that have an expected value of 0. To account for this, let $Y_i = X_i - p$  
$\mathbb{P}(\sum_{i=1}^n (X_i - p) \ge \epsilon) \le e^{-t\epsilon}\Pi_{i=1}^ne^{t^2(b_i-a_i)^2/8} = e^{-t\epsilon}\Pi_{i=1}^ne^{t^2/8} =  e^{nt^2/8-t\epsilon} $ <br>
$\mathbb{P}((\sum_{i=1}^n X_i) - np \ge \epsilon) \le \dots$ <br>
$\mathbb{P}(\bar{X_n} - p \ge \epsilon/n) \le \dots$ <br>
$\mathbb{P}(\mid\bar{X_n} - p\mid \ge \epsilon/n) \le \mathbb{P}(\bar{X_n} - p \ge \epsilon/n) \le \dots$ <br>

Let $\gamma = \epsilon/n \rightarrow n\gamma = \epsilon$ <br>
$\mathbb{P}(\mid\bar{X_n} - p\mid \ge \gamma) \le e^{nt^2/8-tn\gamma} = e^{nt(t/8-\gamma)}$ <br>

Provided $t/8 > \gamma$ and for any $t>0$, Hoeffding's inequality should be smaller for large values of n as it is exponentially decreasing while the Chebyshev bound is inversly decreasing. 


### Exercise 4.4
(a)
Starting from the general Hoeffding's inequality: <br> 
let $Y_i = X_i - p$  
$\mathbb{P}(\sum_{i=1}^n Y_i \ge \epsilon) \le e^{-t\epsilon}\Pi_{i=1}^ne^{t^2(b_i-a_i)^2/8} = e^{nt^2/8-t\epsilon}$ <br> 
$\mathbb{P}(\sum_{i=1}^n (X_i -p) \ge \epsilon) \le \dots$ <br>
$\mathbb{P}(\hat{p_n} - p \ge \epsilon/n) \le \dots$ <br>
$\dfrac{1}{2}\mathbb{P}(\mid\hat{p_n} - p\mid \ge \epsilon/n) \le e^{nt^2/8-t\epsilon}$

Parameter $t$ can be chosen such that hoeffding's bound is minimized for a sum of iid Bernoulli distributed RVs: <br> 
$\dfrac{d}{dt}(e^{nt^2/8-t\epsilon}) = (nt/4-\epsilon)(e^{nt^2/8-t\epsilon}) = 0 \rightarrow t = 4\epsilon/n\quad$ (note: the function's 2nd derivative is always positive, meaning $4\epsilon/n$ is a global minimum)<br>
When $t = 4\epsilon/n$ is plugged back to the previous result, we obtain: <br> 
$\mathbb{P}(\mid\hat{p_n} - p\mid \ge \epsilon/n) \le 2e^{-2\epsilon^2/n}$ <br>
$1 - \mathbb{P}(\mid\hat{p_n} - p\mid \le \epsilon/n) \le 2e^{-2\epsilon^2/n}$ <br>
$\mathbb{P}(\mid\hat{p_n} - p\mid \le \epsilon/n) \ge 1-2e^{-2\epsilon^2/n}$ <br> 

Let $\gamma = \epsilon/n$ <br>
$\mathbb{P}(\mid\hat{p_n} - p\mid \le \gamma) \ge 1-2e^{-2n\gamma^2}$ <br> 

Given $\epsilon_n = \sqrt{log(2/\alpha)/(2n)} \rightarrow \alpha = 2e^{-2n\epsilon_n^2}$
$\mathbb{P}(\mid\hat{p_n} - p\mid \le \epsilon_n) \ge 1 - \alpha$ <br> 


(b)

```python
import numpy as np 
import matplotlib.pyplot as plt


def generate_pn(n,p): 
    sum = 0
    for i in range(0,n):
        if (np.random.rand() < p): 
            sum += 1
    return sum / n

alpha = 0.05
p = 0.4
n = 100
eps  = np.sqrt(np.log(2/alpha)/(2*n))


trials = 10000
X = np.arange(0,trials,1)
Y = np.empty(X.shape)

withinIntegervalCount = 0 
for i in range(0,trials): 
    pn = generate_pn(n,p)
    if (np.abs(pn - p) < eps): 
        withinIntegervalCount += 1
    Y[i] = withinIntegervalCount / (i+1)

plt.figure(1)
plt.plot(X,Y)
plt.xlabel("n")
plt.title("ratio of $\mid\hat{p_n}-p\mid < \epsilon_n$")
plt.show()

```

Result aligns with the expected result. Ratio remains above $1-\alpha$ = 0.95. Although the bound is quite conservative. <br>

```python
import numpy as np
import matplotlib.pyplot as plt
 
def eps(n,alpha): 
    return np.sqrt(np.log(2/alpha)/(2*n))

n_min = 1000
n_max = 10000
N = np.arange(n_min,n_max+1,1)
Y = np.empty(N.shape)

for i in range(0,N.size): 
    Y[i] = eps(N[i], alpha)

plt.figure(1)
plt.plot(N,Y)
plt.title(r'$\epsilon_n(n)$ for $\alpha = 0.05$')
plt.xlabel('n')
plt.ylabel('$\epsilon_n$')
plt.show()

```

Solving $\epsilon_n(\alpha,n)$ for $n$ gives us $n = log(2/\alpha)/(2\epsilon_n^2)$. This results in $n = 2951.10$ for $\alpha = 0.05$ and $\epsilon_n = 0.025$


### Exercise 4.5

Provided z > 0, we have that: <br>
$\mathbb{P}(\mid Z\mid > z) = 2\mathbb{P}(Z > z) $

Also, the definition of $\mathbb{P}(Z > z)$ is: <br>
$\mathbb{P}(Z > z) = \dfrac{1}{\sqrt{2\pi}}\int_{z}^{\infty}e^{-x^2/2}dx$

The integral resembles $erf(x)$ and thus has no analytical solution. Although there exists an inequality which can be used to bound this integral (refer to: https://math.stackexchange.com/questions/414203/integral-inequation). 

Let $\phi(z) = z$ with $z > 0$. Sinze $\phi(z)\ge z$, we have that:

$\phi(z)\int_{z}^{\infty}e^{-x^2/2}dx  \le  \int_{z}^{\infty}\phi(x)e^{-x^2/2}dx$ <br>

$z\int_{z}^{\infty}e^{-x^2/2}dx  \le  \int_{z}^{\infty}xe^{-x^2/2}dx = [-e^{-x^2/2}]_z^{\infty} = e^{-z^2/2}$  <br>

Since z > 0, both sides of the equation can be divided by $z$: <br>

$\int_{z}^{\infty}e^{-x^2/2}dx  \le  \dfrac{e^{-z^2/2}}{z}$  <br>

Finally: <br>
$\mathbb{P}(\mid Z\mid > z) = \dfrac{2}{\sqrt{2\pi}}\int_{z}^{\infty}e^{-x^2/2}dx \le  \dfrac{e^{-z^2/2}}{z} = \sqrt{\dfrac{2}{\pi}}\dfrac{e^{-z^2/2}}{z}$
