### The Empirical Distribution function 


The empirical distribution function $\hat{F}(x)$ is a CDF that increases by $1/n$ in value everytime a data point is crossed. $\hat{F}(x)$ can be obtained given a set of IID RV $X_1,.. X_n \sim F$. More formally: 

$$ \hat{F}_n(x) = \dfrac{\sum_{i=1}^n I(X_i \leq x)}{n}$$

#### Theorems related to the empirical distribution function 

$\mathbb{E}(\hat{F}_n(x)) = F(x)$ <br> 
$\mathbb{V}(\hat{F}_n(x)) = F(x)(1-F(x))/n$ <br> 
$\text{MSE}(\hat{F}_n(x)) = F(x)(1-F(x))/n \rightarrow 0$ <br> 
$\hat{F}_n(x) \xrightarrow[]{P} F(x) $ <br> 

Notice how the right hand side relies on knowing $F(x)$ ahead of time. These expressions are useful to show the convergence and unbiased behaviour of $\hat{F}(x)$. However, the scope of their use seems limited. 

#### Glivenko-Cantelli Theorem 

Let $X_1,...,X_2 \sim F$. Then: <br> 
$$\sup_{x}   \lvert \hat{F}_n(x)- F(x)\rvert \xrightarrow[]{P} 0$$
Now we give an inequality that will be used to constructed a confidence band. 

#### Dvoretzky-Kiefer-Wolfowitz (DKW) Inequality Theorem 
Let $X_1,...,X_n \sim F$. Then, for any $\epsilon > 0$, <br> 

$$\mathbb{P}\big(\sup_{x}\lvert F(x) - \hat{F}_n(x) \rvert > \epsilon \big) \leq 2e^{-2n\epsilon^2}$$

The previous expression can be expressed as a Nonparametric $1-\alpha$ confidence interval:  

Let $L(x) = max\{\hat{F}_n(x) - \epsilon_n, 0\}$ and $U(x) = min\{\hat{F}_n(x) + \epsilon_n, 1\}$ <br>

$1 - \mathbb{P}\big( \sup_{x} \lvert F(x) - \hat{F}_n(x) \rvert \leq  \epsilon \big) \leq 2e^{-2n\epsilon^2} \quad\forall x$ <br> 
$1 - \mathbb{P}\big(\lvert L(x) \leq F(x) \leq U(x) \rvert \leq  \epsilon \big) \leq 2e^{-2n\epsilon^2} \quad\forall x$ <br> 
$ \mathbb{P}\big(\lvert L(x) \leq F(x) \leq U(x) \rvert \leq  \epsilon \big) \geq 1-2e^{-2n\epsilon^2} = 1 - \alpha  \quad\forall x\qquad $ 
Given $\epsilon_n = \sqrt{\dfrac{1}{2n}\text{log}(\dfrac{2}{\alpha})}$


### Statistical Functional 


A statistical function is a function $T$ of a CDF $F$. The mean, variance and median are all statistical functions: <br> 
$\mu = \int xdF(x)$ <br> 
$\sigma^2 = \int (x-\mu)^2dF(x)$ <br> 
$\text{median } m = F^{-1}(1/2)$ <br> 

The <b>Plug-in estimator</b> of $\theta = T(F)$ is defined by $\hat{\theta} = T(\hat{F_n})$

A <b>linear functional</b> is a statistical functional that satisfies the following property: 
$$ T(F) = \int r(x)dF(x)$$


If $T(F) = \int r(x)dF(x)$ for some function r(x) then T is called a <b>linear functional</b>. It can be prove both scalability and superposition are preserved in linear functionals, i.e. $T(aF_1 + bF_2) = aT(F_1) + bT(F_2)$. <br>

#### Plug-in estimator for linear functional theorem
Plug-in estimators are easier to compute for linear functionals: 
$$T(\hat{F}_n) = \int r(x)d\hat{F}_n(x) = \sum_{i=1}^n r(X_i)$$



Calculating $\hat{se}$ is not always straightforward. In many cases, it turns out that $\hat{F}(x) \approx N(F(x),\hat{se}^2)$. In that case, a $1-\alpha$ confidence interval can be obtained for $F(x)$ which is called the <b>Normal-based interval</b> in this book: 
$$ \hat{F}(x)  = F(x) \pm z_{\alpha/2}\hat{se}$$

95% confidence band implies $\alpha = 0.05$


```python
import numpy as np
from scipy.stats import norm
from scipy.stats import cauchy
import matplotlib.pyplot as plt


def normal(): 
    print(np.random.randn())

alpha = 0.05
n = 100  # number of bins 
eps = np.sqrt(1/(2*n)*np.log(2/alpha))    
grid_X = np.linspace(-3,3,n)  # bounds between bins 


bounds_exceeded_count = 0

trial_number = 1000; 
for trial_number in range(0,trial_number): 

    # construct Empirical distribution function F_hat
    N = np.empty(n)
    F_hat = np.empty(N.shape)
    for i in range(0,n):
        #N[i] = np.random.randn()
        N[i] = cauchy.rvs()
    N= np.sort(N)
    i = 0
    for j in range(0,n):  
        while i < n and N[i] < grid_X[j]: 
            i += 1 
        F_hat[j] = i/n
        
    # determine if the bounds were exceeded
    for j in range(0,len(grid_X)):
        # F_x = norm.cdf(grid_X[j]) # F(x) 
        F_x = cauchy.cdf(grid_X[j]) # F(x) 
        if (F_x < F_hat[j]-eps or F_x > F_hat[j]+eps): 
            bounds_exceeded_count += 1
            break

print("Fraction of trials where bounds were exceeded: " + str(round(bounds_exceeded_count/trial_number,4)))
```

Central limit theorem: $\bar{\mu} \rightsquigarrow N(\mu,\sigma^2/n) $

Recall $\hat{F}_n(x) = n^{-1}\sum^n_{i=1}(I(X_i<x))$ <br> 
Letting $Y = I(X_i<x)$ leads to $\bar{Y} =  n^{-1}\sum^n_{i=1}(I(X_i<x)) = \hat{F}_n(x)$. In other words, $\hat{F}_n(x)$ corresponds to empirical mean of $Y$ for which the limiting distribution can be obtained from the central limit theorem: 

$\mu_Y = \mathbb{E}(Y) = \mathbb{E}(I(X_i<x)) = \mathbb{P}(X_i<x) = F(x)$ <br> 
$\sigma^2_Y = \mathbb{V}(Y) = \mathbb{V}(I(X_i<x)) = \mathbb{P}(X_i<x)(1 - \mathbb{P}(X_i<x)) = F(x)(1-F(x))$


$\hat{F}_n(x) \rightsquigarrow N(F(x),F(x)(1-F(x))/n) $

### Exercise 7.7

```python
import urllib
import numpy as np
import matplotlib.pyplot as plt
url = "https://www.stat.cmu.edu/~larry/all-of-statistics/=data/fijiquakes.dat"
file = urllib.request.urlopen(url)
line = file.readline()


# Read the first 1000 data point. 
max_sample_size = 1000
X = np.empty(max_sample_size)
n = 0; 
while line and n < max_sample_size:
    decoded_line = file.readline().decode("utf-8")[:-1]
    X[n] = float(decoded_line.split()[4])
    n+=1
       
X = X[:n]
X = np.sort(X)

F_x = np.empty(max_sample_size)
x_val = np.empty(max_sample_size)

F_x[0] = 0
x_val[0] = X[0]
F_x[1] = 1/n
x_val[1] = X[0]
i = 1
for j in range(1,n): 
    if (X[j] != X[j-1]):
        x_val[i+1] = X[j]
        F_x[i+1] = F_x[i]
        
        F_x[i+2] = F_x[i] + 1/n
        x_val[i+2] = X[j]
        i+=2
    else:
        F_x[i] += 1/n


# compute 95% confidence interval 
alpha = 0.05
eps = np.sqrt(1/(2*n)*np.log(2/alpha))
    
F_x = F_x[:i]
x_val = x_val[:i]    
    
F_lb = F_x - eps
F_ub = F_x + eps
for i in range(0,len(F_x)):
    F_lb[i] = max(0,F_lb[i])
    F_ub[i] = min(1,F_ub[i])

        
plt.figure(1)
plt.title("95% confidence interval around experimental distribution")
plt.plot(x_val, F_x)
plt.plot(x_val, F_lb)
plt.plot(x_val, F_ub)
plt.legend(["Exp. Dist", "Lower Bound", "Upper Bound"])
plt.show()

```

Let $U_1$ be the event $L_{\beta, 4.3}<F(4.3)<U_{\beta, 4.3}$ s.t. $\mathbb{P}(U_1) > 1-\beta$ <br> 
and $U_2$ be the event $L_{\beta, 4.9}<F(4.9)<U_{\beta, 4.9}$ s.t. $\mathbb{P}(U_2) > 1-\beta$<br>

since $U_1$ and $U_2$ are independant events, we have: <br>  
$\mathbb{P}(U_1 \cap U_2) = \mathbb{P}( L_{\beta, 4.9} -U_{\beta, 4.3} \leq F(4.9)-F(4.3) \leq U_{\beta, 4.9} -L_{\beta, 4.3}) = \mathbb{P}(U_1)\mathbb{P}(U_2) > (1-\beta)^2 = 1-\alpha$ <br> 

Since $\alpha = 0.05$ in the problem statement, $\beta = 0.0253$. <br>
Given $n = 1000$, we can compute $\epsilon_{\beta,b}$: <br> 
$\epsilon_{\beta,b} = \sqrt{1/(2n)\text{log}(2/\beta)}= \sqrt{1/(2000)\text{log}(2/0.0253)} = 0.0467$

Which leads to the final $95%$ confidence interval: <br>
$\hat{F}_n(4.3) = 0.276$ and $\hat{F}_n(4.9) = 0.802$ <br> 
$U_{\beta, 4.9} -L_{\beta, \hat{F}_n(4.3)} = min\{1,\hat{F}_n(4.9) + \epsilon_{\beta,n}\} - max\{0,\hat{F}_n(4.3) -\epsilon_{\beta,n}\}  = 0.6194$ <br> 
$L_{\beta, 4.9} -U_{\beta, \hat{F}_n(4.3)} = max\{0,\hat{F}_n(4.9) -\epsilon_{\beta,n}\} - min\{1,\hat{F}_n(4.3) +\epsilon_{\beta,n}\} = 0.4326$ <br>


### Exercise 7.8

Median plug in estimator: $\hat{F}_n^{-1}(1/2)$



```python
import scipy as sci
import numpy as np 
import urllib.request
import matplotlib.pyplot as plt
import scipy.stats

def isDataLine(line):
    if (len(line) == 3 and isNumber(line[0]) and isNumber(line[1]) and isNumber(line[2])):
        return True
    return False

def isNumber(s): 
    return s.isdigit() or s.replace('.','',1).isdigit()

def mean(N):
    sum = 0
    for n in N:
        sum += n
    return sum / len(N)

def variance(N): 
    n = len(N)
    m = mean(N)
    return np.sum(np.power(np.array(N) - m,2))/(n-1) 
     

n = 272
waiting_time = np.empty(n)
eruption_time = np.empty(n)

url = "https://www.stat.cmu.edu/~larry/all-of-statistics/=data/faithful.dat"
file = urllib.request.urlopen(url)

i = 0
for raw_line in file:
    line = raw_line.decode("utf-8")[:-1].split()
    if isDataLine(line): 
        eruption_time[i] =line[1]
        waiting_time[i] = line[2]
        i+= 1

Z_0p90 = scipy.stats.norm.ppf(1 - 0.1/2)
se = np.sqrt(variance(waiting_time)/n) 
mu_et = mean(eruption_time)
mu_wt = mean(waiting_time)
median_et = np.sort(waiting_time)[round(n/2)]
    
print("Mean eruption time: ".ljust(30) + str(round(mu_et,3)))
print("Mean waiting time: ".ljust(30) + str(round(mu_wt,3)))
print("Std error on waiting time: ".ljust(30) + str(round(se,3)))
print("90% confidence interval: ".ljust(30) + "[" + str(round(mu_wt-se*Z_0p90,3)) +", " + str(round(mu_wt+se*Z_0p90,3)) + "]" )
print("Estimated median: ".ljust(30) + str(round(median_et,3)))

```

### Exercise 7.10

```python
import scipy as sci
import numpy as np 
import urllib.request
import matplotlib.pyplot as plt
import scipy.stats

def isDataLine(line):
    if (len(line) == 2 and isNumber(line[0]) and isNumber(line[1])):
        return True
    return False

def isNumber(s): 
    return s.isdigit() or s.replace('.','',1).isdigit()

def mean(N):
    sum = 0
    for n in N:
        sum += n
    return sum / len(N)

def variance(N): 
    n = len(N)
    m = mean(N)
    return np.sum(np.power(np.array(N) - m,2))/(n-1) 
     

n = 26
Unseeded_Clouds	 = np.empty(n)
Seeded_Clouds = np.empty(n)

url = "https://www.stat.cmu.edu/~larry/all-of-statistics/=data/clouds.dat"
file = urllib.request.urlopen(url)

i = 0
for raw_line in file:
    line = raw_line.decode("utf-8")[:-1].split()
    if isDataLine(line): 
        Unseeded_Clouds[i] =line[0]
        Seeded_Clouds[i] = line[1]
        i+= 1

Z_0p95 = scipy.stats.norm.ppf(1 - 0.05/2)
theta_estimate = mean(Unseeded_Clouds) - mean(Seeded_Clouds)
se_unseededClouds = np.sqrt(variance(Unseeded_Clouds)/n) 
se_seededClouds = np.sqrt(variance(Seeded_Clouds)/n) 
se_teta = np.sqrt(se_unseededClouds**2 + se_unseededClouds**2)
    
print("Theta estimate: ".ljust(20) + str(round(theta_estimate,3)))
print("Std error on Theta: ".ljust(20) + str(round(se_teta,3)))
print("95% confidence interval: ".ljust(20) + "[" + str(round(theta_estimate-se_teta*Z_0p95,3)) +", " + str(round(theta_estimate+se_teta*Z_0p95,3)) + "]" )

```
