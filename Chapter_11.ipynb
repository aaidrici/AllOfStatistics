### The Bayesian method: 

How to obtain the posterior distribution: 

$f(\theta|x_1,\dots,x_n) = \dfrac{f(x_1,\dots,x_n,\theta)}{f(x_1,\dots,x_n)} = \dfrac{f(x_1,\dots,x_n|\theta)f(\theta)}{\int f(x_1,\dots,x_n|\theta)f(\theta)d\theta} = \dfrac{\mathcal{L}_n(\theta)f(\theta)}{\int \mathcal{L}_n(\theta)f(\theta)d\theta}= \dfrac{\mathcal{L}_n(\theta)f(\theta)}{c_n} \propto \mathcal{L}_n(\theta)f(\theta)$

where: $\mathcal{L}_n(\theta) = \prod\limits_{i=1}^n f(x_i|\theta)$

For practical use, only proportionality to the posterior density needs to be preserved - all terms that are not function of $\theta$ can be removed. The term that was dropped, $\int \mathcal{L}_n(\theta)f(\theta)d\theta$ is called the <b>normalizing constant</b>. 


$f(\theta|x_1,\dots,x_n)$ is refered to as the <b>posterior</b> <br> 
$\mathcal{L}_n(\theta)$ is refered to as the <b>likelihood</b> <br> 
$f(\theta)$ is refered to as the <b>prior</b>

It is also possible to obtain a Bayesian interval estimate $C=(a,b)$ by selecting $a$ and $b$ such that $\int\limits_{-\infty}^a f(\theta|x^n) = \int\limits_b^\infty f(\theta|x^n) = \alpha/2 $ which results in:  <br> 
$$\mathbb{P}(\theta\in C) = \int\limits_b^a f(\theta|x_1\dots,x_n)d\theta = 1-\alpha$$

When a given prior distribution result in the same posterior distribution, the prior is said to be a <b>conjugates</b> with respect to the model.

```python
import numpy as np 
from scipy.stats import norm

def bayes_method(x, b): 
    n = len(x)
    x = np.sum(x) 
    x2 = np.sum(x**2)
    return np.exp(-x2/2) / (np.exp(-x2/2) + np.exp( x**2/(n+1/b**2)/2 - x2/2)*(n*b**2+1)**(-1/2)  )

def p_val(x,b):
    n = len(x)
    w = sum(x)/np.sqrt(n)
    #return w
    return 2*norm.cdf(-abs(w))

x_arr = np.linspace(-3,3,5)
b_arr = np.array([0.1, 0.2, 0.5, 1, 2, 5])
mu_arr = np.linspace(-0.5,0.5,5)



print("==== PART I: Bayes method vs. Wald's test p-value for sample size 1 ====\n ")

print("Bayes Method: ")
print(9*" "+"|", end="")
for x in x_arr: 
    print(f"x = {x:5.2f},",end = " ")
print()
print("-"*80)
for b in b_arr: 
    print(f"b = {b:5.2f}| ",end = " ")
    for x in x_arr: 
        print(f"{bayes_method([x],b):7.4f},   ", end="")
    print()

print("\n\nWald's test p-value Method: ")
print(9*" "+"|", end="")
for x in x_arr: 
    print(f"x = {x:5.2f},",end = " ")
print()
print("-"*80)
for b in b_arr: 
    print(f"b = {b:5.2f}| ",end = " ")
    for x in x_arr: 
        print(f"{p_val([x],b):7.4f},   ", end="")
    print()


print("\n\n==== PART II: Bayes method vs. Wald's test p-value for sample size n=10 ====\n")


# generate data: 
n = 50
X_arr = [None]*len(mu_arr)
for i in range(0, len(mu_arr)): 
    X_arr[i] = norm.rvs(loc = mu_arr[i], scale = 1, size=n )


print("Bayes Method: ")
print(9*" "+"|", end="")
for mu in mu_arr: 
    print(f"mu = {mu:5.2f},",end = " ")
print()
print("-"*80)
for b in b_arr: 
    print(f"b = {b:5.2f}| ",end = " ")
    for X in X_arr:
        print(f"{bayes_method(X,b):7.4f},   ", end="")
        #print(np.mean(X))
    print()

print("\n\nWald's test p-value Method: ")
print(9*" "+"|", end="")
for mu in mu_arr: 
    print(f"mu = {mu:5.2f},",end = " ")
print()
print("-"*80)
for b in b_arr: 
    print(f"b = {b:5.2f}| ",end = " ")
    for X in X_arr: 
        print(f"{p_val(X,b):7.4f},   ", end="")
    print()

    
```
