### Exercise 16.3

$\theta = \mathbb{E}(C_1) - \mathbb{E}(C_0) = \big(\mathbb{E}(C_1 | X=0)\mathbb{P}(X=0) + \mathbb{E}(C_1 | X=1)\mathbb{P}(X=1)\big) - \big(\mathbb{E}(C_0 | X=0)\mathbb{P}(X=0) + \mathbb{E}(C_0 | X=1)\mathbb{P}(X=1)\big)$

The unobserved random variables must be put a bound to them: <br> 
$0 \leq \mathbb{E}(C_1 | X=0) \leq 1$ <br> 
$0 \leq \mathbb{E}(C_0 | X=1) \leq 1$ <br> 

Therefore: <br>


$
\big(\mathbb{E}(C_1 | X=1)\mathbb{P}(X=1)\big) - \big(\mathbb{E}(C_0 | X=0)\mathbb{P}(X=0) + \mathbb{P}(X=1)\big)
\leq \theta \leq 
\big(\mathbb{P}(X=0) + \mathbb{E}(C_1 | X=1)\mathbb{P}(X=1)\big) - \big(\mathbb{E}(C_0 | X=0)\mathbb{P}(X=0)\big)
$ <br>

$
- \mathbb{P}(X=1)  \leq \theta \leq \mathbb{P}(X=0) 
$ <br>

The width of the bound is: <br> 
$\text{lower bound} - \text{upper bound} = \mathbb{P}(X=0) - \big(-\mathbb{P}(X=1)\big) = \mathbb{P}(X=0) + \mathbb{P}(X=1) = 1 $ 

### Exercise 16.4

$\beta_1$ is a discrete RV that can have values $\beta_{1,1}, \beta_{1,2}, \dots, \beta_{1,k}$ given there are $k$ subjects. Any joint distribution $(X,\beta_1)$ that favours combination of $(x_i,\beta_{1,i})$ where higher value of $x_i$ result in lower values of $\beta_{1,i}$ and vice versa may result in $\mathbb{E}(Y|X=x)$ being a decreasing function of $x$. <br>

The following 3 equations are sufficients conditions to built a joint distribution resultling in $\mathbb{E}(Y|X=x)$: <br>

(1) $\beta_{1,i+1} < \dfrac{x_i}{x_{i+1}}\beta_{1,i}\quad$ (for $x_i < x_{i+1}$)  <br>
(2) $\mathbb{P}(\beta = \beta_{1,i}, X \neq x_{i}) = 0$ <br>
(3) $\mathbb{P}(\beta = \beta_{1,i}, X = x_{i}) > 0 \qquad \forall i $ <br>


Proof: <br>
For a given value $i \in \{1, k\}$ <br>
$\mathbb{E}(Y|X=x_i)$ <br>
$= \mathbb{E}(C_i(x)|X=x_i)$ <br>
$=\mathbb{E}(\beta_{i,1}x_i|X=x_i)$ (from problem statement) <br>
$=\beta_{i,1}x_i$ (applying equation 2 and 3)<br>

Therefore, starting from equation 1 results in: <br> 
$\beta_{1,i+1}x_{i+1} < x_i\beta_{1,i} \implies \mathbb{E}(\beta_{1,i}x_i|X=x_i) < \mathbb{E}(\beta_{1,i+1}x_i|X=x_{i+1}) \implies \mathbb{E}(Y|X=x_i) < \mathbb{E}(Y|X=x_{i+1})\quad$ (for $x_i < x_{i+1}$)  <br>

The logic behind equation 1 is illustrated below. Increments in $x$ must be followed by a sufficient decrement in $\beta_1$ such that $x\beta_1$ can only decrease as $x$ increases:  <br>
$(x_i + \Delta x)\beta_{1,i+1} < x_i\beta_{1,i} \implies x_{i+1}\beta_{1,i+1} < x_i\beta_{1,i} \implies \beta_{1,i+1} < \dfrac{x_i}{x_{i+1}}\beta_{1,i}$
![title](img/Q16_4.PNG)

Equation 2 and 3 are used to simplify the proof when evaluating the term $\mathbb{E}(Y|X=x_i)$. 

Here's an example of joint distribution resulting in a decrease in $\mathbb{E}(Y|X=x)$ as $x$ increases: <br>
$\mathbb{P}(\beta_1 = 1 | x=1) = 1/3$ <br>
$\mathbb{P}(\beta_1 = 1/3 | x=2) = 1/3$ <br>
$\mathbb{P}(\beta_1 = 1/10 | x=3) = 1/3$ <br>
